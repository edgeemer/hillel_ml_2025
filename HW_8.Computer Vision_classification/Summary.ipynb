# Thoughts

## General observations

- All models have high accuracies (as expected from theory & documentation) <= expected because of deep architectures
- Training/Evaluation accuracies difference => overfitting present (especially visible in `DenseNet201`).
- Fluctuations in evaluation loss/accuracy, especially in `DenseNet201` and `EfficientNet_v2_m` => for generalization improvement, if needed, additional techniques can be used, like:
  - dropout
  - learning rate scheduling
  - looking into data augmentation
- Given that our task was to get better knowledge about pre-trained models, it was pretty interesting to test them in real conditions. Despite it was suggested not to use inception models, theory behind them is extremely interesting.
